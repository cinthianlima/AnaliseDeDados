### ii) Soma das variáveis de cada conjunto:

```python
lista_listas = []
for j in range(0, len(L)):
        
    uniform_datas =[]

    for s in L:

        d = []
        for q in range(0, s):
            uniform = np.random.uniform(0,1,n)
            data = uniform
            d.append(data)
        uniform_datas.append(d)
    
    item = uniform_datas[j]
    
    soma_variaveis = []
    for w in range(0, n):
        soma = 0
        for b in range(0, len(item)):
            soma+= item[b][w]
        soma_variaveis.append(soma/L[j])
    
    lista_listas.append(soma_variaveis)
    
    data = Statisticaltools1(soma_variaveis)
    z_m = data.media_aritmetica()
    v_z = data.variance()
    des_z = v_z ** (1/2)
    
    x = np.linspace(z_m - 3*des_z, z_m + 3*des_z, 100)
    gaussiana = gauss(x) #distribuição gaussiana
    
    plt.hist(soma_variaveis, 40, density=True)
    plt.plot(x, gaussiana)
    plt.title('Distribuição Uniforme')
    plt.show()
    
```
